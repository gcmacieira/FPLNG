Encontrar os anos em cada capítulo:

```python
def encontrar_anos(txt):
    # Divide o texto em capítulos usando '#' como marcador de início de capítulo
    capitulos = re.split(r'#', txt)
    
    # Expressão regular para encontrar sequências de 4 números (anos)
    padrao_ano = r'\b1\d{3}\b'
    # Ex: se quissemos só anos entre 1500 e 1999 colocavamos: r'\b1[5-9]\d\d\b'

    for i, capitulo in enumerate(capitulos[1:], start=1):
        anos_encontrados = re.findall(padrao_ano, capitulo)
        if anos_encontrados:
            print(f"Anos encontrados no Capítulo {i}: {anos_encontrados}")

encontrar_anos(txt)
```

**TPC 09/11:**

1- Como aparece um nome próprio em Português (sequência de palavras com letras maiúsculas, podendo haver uns "de" pelo meio);

2- Uma expressão regular que permita encontrar nomes próprios;

3- A partir dessa expressão regular, fazer um "Find All" dos nomes próprios em "A Brasileira de Prazins".

```python
import re

texto = ficheiro

# Removendo quebras de linha
texto_sem_quebras = texto.replace('\n', ' ')


# procura por nomes próprios que começam com uma letra maiúscula, 
# têm mais de três letras, podem ou não ter a palavra "de" no meio (mas não é obrigatória), 
# e terminam com outra palavra que começa com uma letra maiúscula.
padrao = r'\b[A-Z][a-z]{3,}\s?(?:de\s)?[A-Z][a-z]{3,}\b'

# Encontrar todas as correspondências no texto
nomes_encontrados = re.findall(padrao, texto_sem_quebras, flags=re.DOTALL)

# Criar um dicionário para armazenar a contagem de cada nome
contagem_nomes = {}

# Contar a ocorrência de cada nome no texto
for nome in nomes_encontrados:
    contagem_nomes[nome] = contagem_nomes.get(nome, 0) + 1

# Imprimir os resultados, incluindo a contagem
for i, (nome, contagem) in enumerate(contagem_nomes.items(), 1):
    print(f"{i}. {nome} - Aparece {contagem} vez(es) no ficheiro")
```
